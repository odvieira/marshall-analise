```python
import pandas as pd

pd.set_option('display.max_columns', None)
#pd.reset_option('display.max_rows')
pd.set_option('display.max_rows', 20)  # ao inv√©s de none pode por INT do limite
pd.set_option('display.max_colwidth', 15)
```

```python
df = pd.read_csv('../data/df_subvar_9-10_country.csv')
```

```python
df.head()
```

```python
variations = df['SubVariation'].unique().tolist()
variations
```

```python
info = [] 
for variation in variations:
    df_var = df.loc[df['SubVariation'] == variation]
    df_var = df_var.sort_values(by='Date')
    df_var = df_var.iloc[0:10,:]
   

    for index, row in df_var.iterrows():
        linha = {}
        linha['id'] = row['id']
        linha['Variation'] = variation
        linha['Event'] = row['Event']
        linha['Site'] = row['Site']
        linha['Country'] = row['country']
        linha['Year'] = row['Date']
        linha['White'] = row['White']
        linha['Black'] = row['Black']
        linha['Result'] = row['Result']

        if row['Site'] == 'Bargteheide':
            linha['Country'] = 'Germany'
        if row['Site'] == 'Bariloche':
            linha['Country'] = 'Argentina'
        if row['Site'] == 'Bremen ':
            linha['Country'] = 'Germany'
        if row['Site'] == 'Bremer ':
            linha['Country'] = 'Germany'
        if row['Site'] == 'Bruchkoebel':
            linha['Country'] = 'Germany'
        if row['Site'] == 'Budapest':
            linha['Country'] = 'Hungary'
        if row['Site'] == 'Buenos Aires':
            linha['Country'] = 'Argentina'
        if row['Site'] == 'Calcutta ':
            linha['Country'] = 'India'
        if row['Site'] == 'Florida Valle':
            linha['Country'] = 'Argentina'
        if row['Site'] == 'Germany':
            linha['Country'] = 'Germany'
        if row['Site'] == 'HUN':
            linha['Country'] = 'Hungary'
        if row['Site'] == 'Oerebro':
            linha['Country'] = 'Sweden'
        if row['Site'] == 'Parnu':
            linha['Country'] = 'Estonia'
        if row['Site'] == 'Topola ':
            linha['Country'] = 'Serbia'
        if row['Site'] == 'Trencianske Teplice':
            linha['Country'] = 'Slovakia'
        if row['Site'] == 'URS':
            linha['Country'] = 'Russian Federation'
        if row['Site'] == 'USA':
            linha['Country'] = 'United States'
        if row['Site'] == 'Villa Ballester':
            linha['Country'] = 'Argentina'
        if row['Site'] == 'Villa Gesell':
            linha['Country'] = 'Argentina'
        if row['Site'] == 'Villa Gesell':
            linha['Country'] = 'Argentina'

        info.append(linha)

```

```python
variations10 = pd.DataFrame(info)
variations10
```

```python
start_year = []
for variation in variations:
    df_var = variations10.loc[variations10['Variation'] == variation]
    df_var = df_var.sort_values(by='Year')
    info = {}

    info['id'] = df_var['id'].tolist()[0]
    info['Variation'] = variation
    info['FirstOccurrence'] = df_var['Year'].tolist()[0]
    info['Country'] = df_var['Country'].tolist()[0]
    info['White'] = df_var['White'].tolist()[0]
    info['Black'] = df_var['Black'].tolist()[0]
    info['Result'] = df_var['Result'].tolist()[0]
    start_year.append(info)


start = pd.DataFrame(start_year)
```
