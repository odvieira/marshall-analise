```python
import pandas as pd
import locationtagger

pd.set_option('display.max_columns', None)
#pd.reset_option('display.max_rows')
pd.set_option('display.max_rows', 20)  # ao invés de none pode por INT do limite
pd.set_option('display.max_colwidth', 15)
```

```python
df = pd.read_csv('../data/df_subvar_9-10.csv')
```

```python
df = df.loc[(df['WhiteElo'] >= 2500) | (df['WhiteElo'].isna())]
df = df.loc[(df['BlackElo'] >= 2500) | (df['BlackElo'].isna())]

#Escolhe variações que foram jogadas mais de dez vezes
df = df.loc[
    (df['SubVariation'] == '9. d3 dxe4 10. dxe4 Qxd1') |
    (df['SubVariation'] == '9. d4 Nxe4 10. dxe5 Be6') |
    (df['SubVariation'] == '9. d4 exd4 10. e5 Ne4') |
    (df['SubVariation'] == '9. exd5 Nxd5 10. Nxe5 Nxe5') |
    (df['SubVariation'] == '9. exd5 Nxd5 10. d4 exd4') |
    (df['SubVariation'] == '9. exd5 e4 10. Ng5 Bg4') |
    (df['SubVariation'] == '9. exd5 e4 10. dxc6 exf3')
]
```
