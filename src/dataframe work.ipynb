```python
df = pd.read_csv('../data/df_jogos.csv')
```

```python
evals_list = df['Evals'][0]
evals_list = evals_list.strip('[] ').split(",")
print(len(evals_list))
print(evals_list[-1])
print('=============')
print(min(evals_list))
print(max(evals_list))
print(type(evals_list))
evals_list
```

```python

import ast
dados = []

for index, row in df.iterrows():
    linha = {}
    
    # initializing string representation of a list
    evals_list = row['Evals']

    # Converting string to list
    evals_list = ast.literal_eval(evals_list)

    linha['id'] = row['id']
    linha['MaxEval'] = max(evals_list)
    linha['MinEval'] = min(evals_list)
    linha['FinalEval'] = evals_list[-1]

    linha['W<=10'] = 0
    linha['W<=25'] = 0
    linha['W<=50'] = 0
    linha['W<=100'] = 0
    linha['W<=200'] = 0
    linha['W<=500'] = 0
    linha['W>500'] = 0
    linha['B<=10'] = 0
    linha['B<=25'] = 0
    linha['B<=50'] = 0
    linha['B<=100'] = 0
    linha['B<=200'] = 0
    linha['B<=500'] = 0
    linha['B>500'] = 0

    linha['Wprojection'] = 0
    linha['Bprojection'] = 0


    
    turn = False #true branca, false preta
    first = True #primeira jogada
    jogada_anterior = 0
    jogada_atual = 0
    direction = 0 #1 subindo, -1 descendo
    for eval in evals_list:
        turn = not turn
        jogada_atual = eval

        #testa se é o primeiro lance
        if first:
            jogada_anterior = jogada_atual
            first = False
            continue

        #testa de quem é a jogada
        if turn: #branca
            var = abs(jogada_anterior - jogada_atual)
            #testa direção da variação
            if jogada_anterior >= jogada_atual:
                if var <= 10:
                    linha['W<=10'] += 1
                elif var <= 25:
                    linha['W<=25'] += 1
                elif var <= 50:
                    linha['W<=50'] += 1
                elif var <= 100:
                    linha['W<=100'] += 1
                elif var <= 200:
                    linha['W<=200'] += 1
                elif var <= 500:
                    linha['W<=500'] += 1
                else:
                    linha['W>500'] += 1
            else:
                if var <= 10:
                    linha['W<=10'] += 1
                
        else: #preta
            var = abs(jogada_anterior - jogada_atual)
            #testa direção da variação
            if jogada_anterior >= jogada_atual:
                linha['B<=10'] += 1
            else:
                if var <= 10:
                    linha['B<=10'] += 1
                elif var <= 25:
                    linha['B<=25'] += 1
                elif var <= 50:
                    linha['B<=50'] += 1
                elif var <= 100:
                    linha['B<=100'] += 1
                elif var <= 200:
                    linha['B<=200'] += 1
                elif var <= 500:
                    linha['B<=500'] += 1
                else:
                    linha['B>500'] += 1

        #Projeção das evaluation
        if jogada_atual > 0:
            if jogada_atual < 1000:
                linha['Wprojection'] += jogada_atual
            else:
                linha['Wprojection'] += 1000
        else:
            if jogada_atual > -1000:
                linha['Bprojection'] += jogada_atual
            else:
                linha['Bprojection'] += -1000
        
        jogada_anterior = jogada_atual

    linha['DifProjection'] = linha['Wprojection'] + linha['Bprojection']
    dados.append(linha)

df_evals = pd.DataFrame(dados)
df_evals.to_csv('../data/df_evals.csv', index=False)
        
```
