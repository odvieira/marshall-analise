```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
#pd.reset_option('display.max_rows')
pd.set_option('display.max_rows', 20)  # ao invés de none pode por INT do limite
pd.set_option('display.max_colwidth', 15)
```

```python
df = pd.read_csv('../data/df_subvar_9-10.csv')
```

```python
df_players = pd.read_csv('../data/players_list_fide.csv', sep=';')
df_players.head()
```

```python
russ_list = [
'Armenia',
'Azerbaijan',
'Belarus',
'Estonia',
'Georgia',
'Kazakhstan',
'Kyrgyzstan',
'Latvia',
'Lithuania',
'Moldova',
'Russia',
'Tajikistan',
'Turkmenistan',
'Ukraine',
'Uzbekistan']
```

```python
df_fide = pd.read_csv('../data/fide_feds.csv', sep=';')
df_fide.head()
```

```python
count = -1
for index, row in df.iterrows():
    count += 1
    print('{0}\t\t'.format(count), end='\r')
    white = row['White']
    black = row['Black']

    wname = df_players.loc[df_players['Name'] == white]['Fed'].tolist()
    bname = df_players.loc[df_players['Name'] == black]['Fed'].tolist()

    if len(wname) == 0:
        wname = np.NaN
        wfed = np.NaN
    else:
        wname = wname[0]
        wfed = df_fide.loc[df_fide['fed'] == wname]['continent'].tolist()[0]
        wcountry = df_fide.loc[df_fide['fed'] == wname]['country'].tolist()[0]

        if wcountry in russ_list:
            wfed = 'URSS'


    if len(bname) == 0:
        bname = np.NaN
        bfed = np.NaN
    else:
        bname = bname[0]
        bfed = df_fide.loc[df_fide['fed'] == bname]['continent'].tolist()[0]
        bcountry = df_fide.loc[df_fide['fed'] == bname]['country'].tolist()[0]

        if bcountry in russ_list:
            wfed = 'URSS'

    df.loc[index, 'WhiteRegion'] = wfed
    df.loc[index, 'BlackRegion'] = bfed

```

```python
copia = df.copy()
```

```python
df = copia.copy()
```

```python
#df = df.loc[(df['WhiteElo'] >= 2000) | (df['WhiteElo'].isna())]
#df = df.loc[(df['BlackElo'] >= 2000) | (df['BlackElo'].isna())]
```

```python
df['WGoodMoves'] = df['W<=10'] + df['W<=25'] + df['W<=50']
df['BGoodMoves'] = df['B<=10'] + df['B<=25'] + df['B<=50']
df['WBadMoves'] = df['W<=100'] + df['W<=200'] + df['W<=500'] + df['W>500']
df['BBadMoves'] = df['B<=100'] + df['B<=200'] + df['B<=500'] + df['B>500']
df.columns
		
```

```python
df = df[['Date', 'BlackElo', 'WhiteElo', 'SubVariation', 'WGoodMoves','BGoodMoves', 'WBadMoves', 'BBadMoves', 'WhiteRegion', 'BlackRegion']]
```

```python
df
```

```python
# Separar nas sete variações mais comuns e nas demais
common_variations = ['9. d3 dxe4 10. dxe4 Qxd1',
'9. d4 Nxe4 10. dxe5 Be6',
'9. d4 exd4 10. e5 Ne4',
'9. exd5 Nxd5 10. Nxe5 Nxe5',
'9. exd5 Nxd5 10. d4 exd4',
'9. exd5 e4 10. Ng5 Bg4',
'9. exd5 e4 10. dxc6 exf3']

for index, row in df.iterrows():
    if row['SubVariation'] not in common_variations:
        df.loc[index, 'SubVariation'] = 'Other'
```

# Regressão Linear

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
import numpy as np

from stargazer.stargazer import Stargazer #visualizar diferentes modelos
```

```python
model = smf.ols(formula="Q('WGoodMoves') ~  Q('Date') + Q('BlackElo') + Q('WhiteElo') + \
       C(SubVariation) + C(WhiteRegion) + C(BlackRegion)", data=df)

res_Wgm = model.fit()
#res_Wgm.summary()
```

```python
model = smf.ols(formula="Q('WBadMoves') ~  Q('Date') + Q('BlackElo') + Q('WhiteElo') + \
         C(SubVariation) + C(WhiteRegion) + C(BlackRegion)", data=df)

res_Wbm = model.fit()
#res_Wbm.summary()
```

```python
model = smf.ols(formula="Q('BGoodMoves') ~  Q('Date') + Q('BlackElo') + Q('WhiteElo') + \
        C(SubVariation) + C(WhiteRegion) + C(BlackRegion)", data=df)

res_Bgm = model.fit()
#res_Bgm.summary()
```

```python
model = smf.ols(formula="Q('BBadMoves') ~  Q('Date') + Q('BlackElo') + Q('WhiteElo') + \
        C(SubVariation) + C(WhiteRegion) + C(BlackRegion)", data=df)

res_Bbm = model.fit()
#res_Bbm.summary()
```

```python
stargazer = Stargazer([res_Wgm, res_Wbm, res_Bgm, res_Bbm])
stargazer
```
