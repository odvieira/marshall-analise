```python
import pandas as pd
```

```python
df = pd.read_csv('../data/df_jogos.csv')
df
```

```python
def string_analyzer(row):
    s = row['Site']
    splitted_s = s.split(' ')
    first_part = splitted_s[:-1]
    last_part = splitted_s[-1]
    city = ''

    if last_part.isupper() and len(splitted_s) > 1:
        for w in first_part:
            city += w + ' '
    else:
        city = s

    row['Site'] = city

    return row
```

```python
aux = df.apply(lambda row: string_analyzer(row=row), axis=1)
```

```python
aux.to_csv('../data')
```
